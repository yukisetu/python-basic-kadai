```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

human = Human("やま", 77)

print(human.name)
print(human.age)

```

```python
class Human:
  def __init__(self):
    #属性を定義
    self.name = ""
    self.age = ""

name = Human()
age = Human()

name.name = "やま"
age.age = 77

print(name.name)
print(age.age)
```
