```python
class Human:
  def __init__(self, name, age):
    # 属性の定義
    self.name = name
    self.age = age

  def check_adult(self):
      if self.age >= 20:
        print("{} は大人".format(self.name))

      else:
        print("{} は子供".format(self.name))

# インスタンスの作成
human = [["やま", 77],
           ["ゆう", 7]]

#インスタンスのリストへ追加
human_list = []
for a in human:
  human_list.append((a[0], a[1]))

#メゾット
i = 0
for b in human_list:
  Human(b[0], b[1]).check_adult()
```
